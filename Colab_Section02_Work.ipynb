```python
import pandas as pd

url = "https://raw.githubusercontent.com/AmilaDarshanaJ/bookstore-data-analysis/refs/heads/main/bookstore_transactions.csv"
df = pd.read_csv(url)
df.head()
```

```python
# Preview and confirm shape
print(df.shape)
print(df.head())

# Check for duplicates (good practice)
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Check for missing values
print(df.isnull().sum())

# Optionally fill or drop
df = df.fillna(0)  # or df.dropna() if appropriate
```

```python
import numpy as np

# Example on numeric columns only
print("Mean:\n", df.mean(numeric_only=True))
print("Median:\n", df.median(numeric_only=True))
print("Standard Deviation:\n", df.std(numeric_only=True))
print("Minimum:\n", df.min(numeric_only=True))
print("Maximum:\n", df.max(numeric_only=True))
print("Quantiles:\n", df.quantile([0.25, 0.5, 0.75], numeric_only=True))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(df.select_dtypes(include='number'))
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
